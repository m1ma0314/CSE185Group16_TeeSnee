```python
cellxgene_mat= "./data/X.csv"
output = "./"
target_perplexity = 100
clusters = teesnee.calculate_tSNE(cellxgene_mat, output, target_perplexity)
```

```python
clusters
```

```python
import pandas as pd
my_dataset = pd.read_csv(cellxgene_mat)
my_dataset["clusters"] = clusters
my_dataset
```

```python
my_dataset.loc[my_dataset["clusters"] == 0]
```

```python
min_dataset = pd.read_csv("./data/X.csv")
```

```python
num_samples = list(range(10,len(min_dataset)))
```

```python
import time 
cpu_times = []
for i in range(10, len(min_dataset)):
    curr_dataset = min_dataset[0:i]
    start_time = time.time()  # Get the start time
    curr_dataset.to_csv('./curr_dataset.csv', index=False, header=True)
    teesnee.calculate_tSNE('./curr_dataset.csv', output, target_perplexity)
    end_time = time.time()  # Get the end time
    cpu_time = (end_time - start_time) * 1000
    cpu_times.append(cpu_time)
```

```python
import matplotlib.pyplot as plt

# Assuming you have loaded the data into two arrays: num_samples and cpu_time

plt.plot(num_samples, cpu_times, 'o-')  # 'o-' represents a line plot with markers
plt.xlabel('Number of Samples')
plt.ylabel('CPU Time (ms)')
plt.title('CPU Time vs Number of Samples')
plt.grid(True)
plt.show()
```

# scanpy's t-sne

```python
import scanpy as sc
import pandas as pd
```

```python
%%time
data = pd.read_csv('./data/X.csv')
adata = sc.AnnData(data.values)

# adata = sc.read('minimal_dataset.csv')
sc.pp.neighbors(adata)
# Perform t-SNE embedding

sc.tl.tsne(adata)
sc.tl.leiden(adata, key_added='leiden_clusters', resolution=0.5)
sc.pl.tsne(adata, color='leiden_clusters', size=50)
```

```python
sc.tl.leiden(adata, key_added='leiden_clusters', resolution=0.3)
sc.pl.tsne(adata, color='leiden_clusters', size=50)
```
